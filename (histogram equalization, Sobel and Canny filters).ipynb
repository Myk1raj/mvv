```python
## Task 1: Implement Histogram Equalization 
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def load_image_22bai1118(filename_22bai1118):
    """Load the grayscale image and convert it to a numpy array."""
    image_22bai1118 = Image.open(filename_22bai1118).convert('L')
    return np.array(image_22bai1118)

def compute_histogram_22bai1118(image_array_22bai1118):
    """Compute histogram of the image."""
    histogram_22bai1118, bins_22bai1118 = np.histogram(image_array_22bai1118.flatten(), bins=256, range=[0, 256])
    return histogram_22bai1118, bins_22bai1118

def compute_cdf_22bai1118(histogram_22bai1118):
    """Compute the Cumulative Distribution Function (CDF) and normalize it."""
    cdf_22bai1118 = histogram_22bai1118.cumsum()
    cdf_normalized_22bai1118 = cdf_22bai1118 * 255 / cdf_22bai1118[-1]  
    return cdf_normalized_22bai1118

def apply_histogram_equalization_22bai1118(image_array_22bai1118, bins_22bai1118, cdf_normalized_22bai1118):
    """Apply histogram equalization on the image."""
    return np.interp(image_array_22bai1118.flatten(), bins_22bai1118[:-1], cdf_normalized_22bai1118).reshape(image_array_22bai1118.shape)

def plot_results_22bai1118(image_array_22bai1118, image_equalized_22bai1118):
    """Display the original and equalized images along with their histograms."""
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 2, 1)
    plt.imshow(image_array_22bai1118, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(2, 2, 2)
    plt.imshow(image_equalized_22bai1118, cmap='gray')
    plt.title("Equalized Image")
    plt.axis("off")

    plt.subplot(2, 2, 3)
    plt.hist(image_array_22bai1118.flatten(), bins=256, color='blue', alpha=0.7)
    plt.title("Histogram of Original Image")

    plt.subplot(2, 2, 4)
    plt.hist(image_equalized_22bai1118.flatten(), bins=256, color='red', alpha=0.7)
    plt.title("Histogram of Equalized Image")

    plt.show()

# Main execution
filename_22bai1118 = 'input.jpg'
image_array_22bai1118 = load_image_22bai1118(filename_22bai1118)
histogram_22bai1118, bins_22bai1118 = compute_histogram_22bai1118(image_array_22bai1118)
cdf_normalized_22bai1118 = compute_cdf_22bai1118(histogram_22bai1118)
image_equalized_22bai1118 = apply_histogram_equalization_22bai1118(image_array_22bai1118, bins_22bai1118, cdf_normalized_22bai1118)

plot_results_22bai1118(image_array_22bai1118, image_equalized_22bai1118)

## Task 2: Compare Sobel and Canny Edge Detection Methods 
# Import necessary libraries
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt

def load_image_22bai1118(filename_22bai1118):
    """Load a grayscale image."""
    return cv2.imread(filename_22bai1118, cv2.IMREAD_GRAYSCALE)

def sobel_edge_detection_22bai1118(image_22bai1118):
    """Apply Sobel edge detection."""
    sobel_x_22bai1118 = cv2.Sobel(image_22bai1118, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y_22bai1118 = cv2.Sobel(image_22bai1118, cv2.CV_64F, 0, 1, ksize=3)
    sobel_edges_22bai1118 = cv2.magnitude(sobel_x_22bai1118, sobel_y_22bai1118)
    return np.uint8(sobel_edges_22bai1118)

def canny_edge_detection_22bai1118(image_22bai1118):
    """Apply Canny edge detection."""
    return cv2.Canny(image_22bai1118, 100, 200)

def measure_execution_time_22bai1118(func_22bai1118, image_22bai1118):
    """Measure execution time for an edge detection function."""
    start_time_22bai1118 = time.time()
    result_22bai1118 = func_22bai1118(image_22bai1118)
    execution_time_22bai1118 = time.time() - start_time_22bai1118
    return result_22bai1118, execution_time_22bai1118

def plot_results_22bai1118(images_22bai1118, titles_22bai1118):
    """Display the images with their respective titles."""
    fig_22bai1118, axes_22bai1118 = plt.subplots(3, 4, figsize=(16, 10))

    for i_22bai1118, ax_22bai1118 in enumerate(axes_22bai1118.flatten()):
        if i_22bai1118 >= len(images_22bai1118):
            ax_22bai1118.axis('off')
            continue
        ax_22bai1118.imshow(images_22bai1118[i_22bai1118], cmap='gray')
        ax_22bai1118.set_title(titles_22bai1118[i_22bai1118])
        ax_22bai1118.axis('off')

    plt.tight_layout()
    plt.show()

# Load images
simple_image_22bai1118 = load_image_22bai1118('Simple.png')
complex_image_22bai1118 = load_image_22bai1118('Complex.png')

# Apply edge detection with timing
sobel_simple_22bai1118, time_sobel_simple_22bai1118 = measure_execution_time_22bai1118(sobel_edge_detection_22bai1118, simple_image_22bai1118)
sobel_complex_22bai1118, time_sobel_complex_22bai1118 = measure_execution_time_22bai1118(sobel_edge_detection_22bai1118, complex_image_22bai1118)
canny_simple_22bai1118, time_canny_simple_22bai1118 = measure_execution_time_22bai1118(canny_edge_detection_22bai1118, simple_image_22bai1118)
canny_complex_22bai1118, time_canny_complex_22bai1118 = measure_execution_time_22bai1118(canny_edge_detection_22bai1118, complex_image_22bai1118)

# Prepare images and titles for display
images_22bai1118 = [
    simple_image_22bai1118, sobel_simple_22bai1118, canny_simple_22bai1118, 
    complex_image_22bai1118, sobel_complex_22bai1118, canny_complex_22bai1118
]

titles_22bai1118 = [
    "Original Simple Image", f"Sobel Simple ({time_sobel_simple_22bai1118:.4f}s)", f"Canny Simple ({time_canny_simple_22bai1118:.4f}s)",
    "Original Complex Image", f"Sobel Complex ({time_sobel_complex_22bai1118:.4f}s)", f"Canny Complex ({time_canny_complex_22bai1118:.4f}s)"
]

# Display results
plot_results_22bai1118(images_22bai1118, titles_22bai1118)
```
